# Instalación

Estos notebooks están pensados para ser usados en Google Colab. Para esto deben iniciar sesion con su cuenta de google (puede ser su personal o la de la universidad) y luego ir a Google Colab https://colab.research.google.com donde aparecerá una pestaña para abrir algún archivo. Deben ir a la pestaña que dice GitHub, donde copiaran la URL del repositorio donde están alojados los notebooks y presionan buscar. Apareceran los distintos notebooks del curso, y solo tienen que elegir uno para abrirlo. Para poder hacer uso de los notebooks, se debe agregar el string que se compartirá por ucursos en la sección "Configuración y conexión al servidor" al inicio de cada notebook.



En caso de algun error al momento de abrirlos o luego de abiertos algún problema con la ejecución del los bloques de código por favor contáctenme por cualquiera de los metodos que aparecen más abajo en la sección Feedback.
