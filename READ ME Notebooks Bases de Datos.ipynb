# Instalación

Estos notebooks estan pensados para ser usados en Google Colab. Para esto deben iniciar sesion con su cuenta de google (puede ser su personal o la de la universidad) y luego ir a Google Colab https://colab.research.google.com donde aparecera una pestaña para abrir algun archivo. Deben ir a la pestaña que dice GitHub, donde copiaran la URL del repositorio donde estan alojados los notebooks y presionan buscar. Apareceran los distintos notebooks del curso, y solo tienen que elegir uno para abrirlo. Para poder hacer uso de los notebooks, se debe agregar el string que se compartira por ucursos en la seccion "Configuracion y conexion al servidor" al inicio de cada notebook.

# Como usar los notebooks
